### Metrics

#### Regression
To evaluate how good we are doing on the **regression** task we will use the Root Mean Squared Error (RMSE). RMSE is given by

$$
\sqrt{\frac{1}{n}\sum\limits_{i=1}^{n}{\Big(d_i -f_i\Big)^2}}
$$


where:

* $n$ is the number of test samples
* $d_i$ is the ground truth value of the i-th sample
* $f_i$ is the predicted value of the i-th sample


#### Classification
To evaluate how good we are doing on the **classification** task we will use the accuracy metrics. Accuracy is given by

$$
\frac{TP+TN}{TP+TN+FP+FN}
$$

where:

* TP is the number of *correctly* classified positive samples
* TN is the number of *correctly* classified negative samples
* FP is the number of *incorrectly* classified positive samples
* FN is the number of *incorrectly* classified negative samples

Ridefinizione della feature target trasformandola in booleana.
Le seguenti 3 funzioni sono:


-**boolean_feature**, prende una colonna e un dataframe e valuta se questa colonna è composta da soli booleani;


-**transform_f2b**, prende un dataframe e trasforma ogni occorrenza di colonne booleane (di tipo float dagli elementi 0 e 1) in boolean, il suo utilizzo comporta una più grande domain knowledge e accresce di conseguenza le prestazioni del modello.


-**modify_shares** trasforma la colonna target da valori continui a valori booleani (utilizzata per la classificazione).

```python
def boolean_feature(colonna,dataframe):
  # Funzione per verificare se una colonna contiene solo valori booleani (0 e 1)
  for elem in dataframe.itertuples(index=False):
    if elem[colonna]!=1 and elem[colonna]!=0:
      return False
  return True
def transform_f2b(examples,execute=True):
  # Funzione per trasformare colonne contenenti solo valori booleani in tipo di dato booleano
  if execute==False:
    return
  colonne_bool=[]
  for colonna in range(len(examples.columns)):
    if boolean_feature(colonna,examples):
      colonne_bool.append(examples.columns[colonna])
  examples[colonne_bool] = (examples[colonne_bool]).astype(bool)
  return examples
def modify_shares(y,execute=True):
  # Funzione per modificare i valori della colonna "shares" in base a una condizione
  if execute==False:
    return y
  list=[]
  for element in y.iteritems():
    if element[1]>=1400:
      list.append(1)
    else:
      list.append(0)
  return pd.DataFrame(list,columns=["shares"])
```

Definizione della struttura dati albero (binario) ed eventuali metodi necessari.

Le performance come si potrà vedere sono equiparabili a quelle di sklearn per quanto riguarda la accuracy, possibili ottimizzazioni potrebbero derivare dal limitare altezza/numero di nodi (massimi) dell'albero da input (es. Sklearn).

L'implementazione si basa su due classi, una classe Nodo caratterizzata da figlio sinistro e destro, lista di features, valori di split, feature scelta per lo split e classificazione nel caso in cui sia una foglia (posta a None altrimenti) mentre la classe DecisionTree è caratterizzata dalla altezza massima selezionabile da input, di default posta a np.inf, e dal nodo sorgente.

In fase di costruzione l'albero utilizza come metodo di discretizzazione la mediana, come consigliato a lezione. Tale discretizzazione viene effettuata ad ogni feature valutata e su cui viene effettuata una split, a sinistra andranno tutti i sample rimanenti che hanno tale feature minore o uguale della mediana, a destra maggiore.

Una volta costruito l'albero seguendo quanto fatto a lezione (utilizzando dunque il concetto di **binary cross-entropy** e di remainder) il metodo predict non fa altro che far percorrere ai diversi sample in input l'intero albero dalla radice ad una delle foglie, avendo tali foglie una classificazione, ottenuta mediante majority voting o mediante same_classification, arrivato ad una foglia il sample verrà classificato con la classificazione di quel nodo dell'albero.

```python
class Nodo:
  def __init__(self,samples,features,left,right,classification,value,feature):
    self.samples=samples #un dataframe
    self.features=features #una lista, da vedere
    self.left=left #un nodo
    self.right=right #un nodo
    self.classification=classification #un valore tra 0 e 1 (low o high)
    self.value=value #valore di split (mediana)
    self.feature=feature
class DecisionTree:
  def __init__(self,altezza_max=np.inf):
    #costruttore del Decision Tree
    if altezza_max<=0:
      raise Exception("ALTEZZA ALMENO 1")
    self.altezza_max=altezza_max
    self.root=None
  def __Learn(self,samples,outputs,features,parent_samples,altezza):
    # Metodo privato per apprendere l'albero decisionale ricorsivamente
    if samples.empty:
      return self.plurality_value(parent_samples,features,outputs)
    elif altezza==self.altezza_max:
      return self.plurality_value(samples,features,outputs)
    elif self.same_classification(outputs):
      return Nodo(samples, features,None, None,outputs.iloc[0,0],None,None)
    elif len(features)==0:
      return self.plurality_value(samples,None,outputs)
    else:
      A,value=self.Importance(samples,features,outputs)
      exs_left=samples.loc[samples[A]<=value,:]
      exs_right = samples.loc[samples[A] > value, :]
      copy_features=features.copy()
      copy_features.remove(A)
      left_tree=self.__Learn(exs_left,outputs.loc[samples[A] <= value],copy_features,samples,altezza+1)
      right_tree=self.__Learn(exs_right,outputs.loc[samples[A]>value],copy_features,samples,altezza+1)
      return Nodo(samples, features, left_tree, right_tree, None, value, A)

  def same_classification(self,outputs):
    # Verifica se tutte le etichette di output hanno la stessa classificazione
    test_value=outputs.iloc[0,0]
    for element in outputs.itertuples():
      if element[1]!=test_value:
        return False
    return True

  def plurality_value(self,samples,features,outputs):
    # Determina il valore di classificazione più frequente nell'insieme di samples rimanenti
    n_positive=0
    n_negative=0
    for element in outputs.itertuples():
      if element[1]==1:
        n_positive+=1
      else:
        n_negative+=1
    return Nodo(samples, features, None, None,1,None,None) if n_positive>n_negative else Nodo(samples, features, None, None,0,None,None)

  def Importance(self,samples,features,outputs):
    # Calcola l'importanza di ogni feature e restituisce la migliore
    entropy_all=self.entropy(outputs)
    max_gain=-np.inf
    winning_feature=None
    mediana4feature=None
    for feature in features:
        if samples[feature].dtype==bool:
          left_subset = outputs.loc[samples[feature] == False]
          right_subset = outputs.loc[samples[feature] == True]
          median=False
        else:
          median=self.median_split(samples,feature)
          left_subset = outputs.loc[samples[feature] <= median]
          right_subset = outputs.loc[samples[feature] > median]
        remainder=self.remainder(right_subset,left_subset,len(outputs)) #questa chiamata va assolutamente corretta
        gain=entropy_all-remainder
        if gain>max_gain:
          max_gain=gain
          winning_feature=feature
          mediana4feature=median
    return (winning_feature,mediana4feature)


  def entropy(self, outputs):
    # Calcola l'entropia di un insieme di etichette di output
    if outputs.empty:
      return 0
    else:
      conteggio_positivi=0
      for y in outputs.itertuples():
        if y[1]==1:
          conteggio_positivi+=1
      p=conteggio_positivi/len(outputs)
      if p==0 or p==1:
        return 0
      else:
        return -(p*math.log2(p)+(1-p)*math.log2(1-p))

  def remainder(self, right_subset,left_subset,n_elements):
    # Calcola il remainder di un attributo dopo uno split
    sum=len(right_subset)*self.entropy(right_subset)+len(left_subset)*self.entropy(left_subset)
    return sum/n_elements

  def median_split(self,samples,feature):
    # Calcola la mediana per una feature
    return statistics.median(samples.loc[:,feature])

  def __predict(self,sample,Node):
    # Metodo privato per effettuare la previsione per un singolo campione
    if Node.value is not None:
      if sample[Node.feature]<=Node.value:
        return self.__predict(sample,Node.left)
      else:
        return self.__predict(sample,Node.right)
    else:
      return Node.classification
  def predict_classification(self,samples):
    # Prevedi le classificazioni per un insieme di campioni
    lista=[]
    for index in range(len(samples)):
      valore=self.__predict(samples.iloc[index],self.root)
      lista.append(valore)
    return pd.DataFrame(lista,columns=["shares"])

  def learn_classification(self,samples,outputs,features,parent_samples):
    # Apprendi l'albero decisionale per la classificazione
    self.root=self.__Learn(samples,outputs,features,parent_samples,0)
    return self.root

def print_Node(nodo):
  # Stampa ricorsivamente il contenuto di un nodo e dei suoi sottoalberi
  if nodo is None:
    #do nothing
    return
  else:
    print("classificazione: ",str(nodo.classification),"feature: ",str(nodo.feature), "valore di split: ",str(nodo.value))
    print_Node(nodo.left)
    print_Node(nodo.right)
def altezza(nodo):
  # Calcola l'altezza dell'albero a partire da un nodo
  if nodo is None:
    return -1
  else:
    return max(altezza(nodo.left),altezza(nodo.right))+1
def n_nodi(nodo):
  # Calcola il numero di nodi dell'albero a partire da un nodo
  if nodo is None:
    return 0
  else:
    return n_nodi(nodo.left)+n_nodi(nodo.right)+1
```

Implementazione regressione lineare (no reg, L1, L2 ed elastic net) con la **possibilità di cambiare da input la task tra classificazione e regressione per il modello**. Ispirandomi alla implementazione di sklearn ho fornito la possibilità di specificare in fase di costruzione del modello diversi parametri:

1) **la tolleranza**, ovvero quel valore minimo di cambiamento che vi deve essere tra due diverse iterazioni nella applicazione della gradient descent, nel caso in cui la variazione sia più bassa della tolleranza si assume l'algoritmo abbia converso.

2) il numero massimo di iterazioni, **max_iter**, prima che l'algoritmo di training termini forzatamente assumendo di essersi avvicinato opportunatamente al minimo globale (che per la funzione di loss MSE sappiamo per certo esistere essendo questa convessa).

3) **alpha**, il learning rate che determina la velocità di convergenza dell'algoritmo

4) **lambda**, lo scalare che determina l'influenza della regolarizzazione

5) **reg**, che assume valori None, "l1", "l2" ed "elastic-net" per specificare l'utilizzo di regolarizzazione per evitare l'overfitting

6) la **task** che naturalmente assume valori "regression" e "classification".

Naturalmente, come spesso avviene nel caso di gradient descent, è stato necessario normalizzare i dati in input usando lo scaler di sklearn.

La differenza principale tra l'esecuzione del modello per la **task di classificazione** e quella per la **task di regressione** è l'applicazione della funzione soglia (threshold a 0.5) con l'update dei pesi mediante l'utilizzo della tecnica di Perceptron nel caso della classificazione. Nel caso della regressione si esegue direttamente la tecnica del Gradient Descent.

```python
class LinearR:
  def __init__(self, tol=0.0001,max_iter=10000,alpha=0.001,l=0.1,reg=None,task="regression"):
    self.tol=tol # Tolleranza per la convergenza dell'algoritmo di discesa del gradiente
    self.max_iter=max_iter # Numero massimo di iterazioni per l'algoritmo di discesa del gradiente
    self.alpha=alpha # Tasso di apprendimento (learning rate)
    self.weights=None # Pesi del modello
    self.l=l # Fattore di regolarizzazione
    if task is not None and task not in {"regression","classification"}:
      raise Exception("Inserire valore di task ammissibile: regression o classification")
    self.task=task  # Task del modello (regressione o classificazione)
    if reg is not None and reg not in {"l1","l2","elastic-net"}:
      raise Exception("Inserire valore di reg ammissibile: l1, l2, elastic-net o None")
    self.reg=reg # regolarizzazione da applicare tra l1, l2, elastic-net o nessuna
  def __Gradient_Descent(self,x,y):
    self.weights=np.zeros(x.shape[1]) # Inizializzazione dei pesi a zero
    for i in range(self.max_iter):
      if self.task=="regression":
        h_x=np.dot(x,self.weights)  # Calcolo dell'output del modello per la regressione
      else:
        h_x=np.where(np.dot(x,self.weights)>=0,1,0) # Calcolo dell'output del modello per la classificazione
      residui=y-h_x  # Calcolo dei residui
      gradiente=np.dot(x.T,residui) # Calcolo del gradiente
      update=2*self.alpha/len(y)*(gradiente) # Calcolo dell'aggiornamento dei pesi

      if self.reg=="l2": #se regolarizzazione l2
        update=self.alpha*self.l*(2*gradiente/len(y)+2*self.l*self.weights)
        self.weights+=update
      elif self.reg is None: #se nessuna regolarizzazione
         self.weights+=update
      elif self.reg=="l1": #se regolarizzazione l1
        sign_weights = np.sign(self.weights)
        update = self.alpha * self.l * (2*gradiente / len(y) + self.l * sign_weights)
        self.weights+=update
      elif self.reg=="elastic-net": #se regolarizzazione elastic-net
        sign_weights = np.sign(self.weights)
        update = self.alpha * self.l * (2*gradiente / len(y) + self.l * sign_weights) + self.alpha*self.l*(2*gradiente/len(y)+2*self.l*self.weights)
        self.weights+=update
      if (update<self.tol).all(): #condizione di convergenza per il modello
        print("HO CONVERSO")
        return
  def fit(self, x, y):
    scaler = StandardScaler()
    x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
    x_array=x.to_numpy() # Standardizzazione delle features
    x_array=np.insert(x_array,0,1,axis=1) # Aggiunta della colonna di bias (intercetta) al dataset di input
    if isinstance(y,pd.DataFrame):
      y=y["shares"]
    y_array=y.to_numpy()
    self.__Gradient_Descent(x_array,y_array) # Apprendimento dei pesi del modello utilizzando Gradient Descent

  def predict(self,x):
    scaler = StandardScaler()
    x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
    x_array=x.to_numpy()
    x_array=np.insert(x_array,0,1,axis=1)  # Aggiunta della colonna di bias (intercetta) al dataset di input
    if self.weights is None:
      raise Exception("Il modello va prima allenato")
    if self.task=="regression":
      pred=np.dot(x_array,self.weights) # Calcolo delle previsioni del modello per la regressione
      return pred
    else:
      return np.where(np.dot(x_array,self.weights)>=0,1,0)  # Calcolo delle previsioni del modello per la classificazione
```

Implementazione Regressione Logistica.
Come per la Linear regression ho fornito la possibilità di modificare da input 3 parametri:
- **Tolleranza**
-**max_iter**
-**learning rate**.


Alcune scelte effettuate durante l'implementazione sono state:

-**inizializzazione dei pesi a 0**

-**criterio di convergenza**: se l'update effettuato su tutti i pesi è stato **per ogni peso** MINORE della tolleranza, allora l'algoritmo converge

-**normalizzazione dei dati in input**, senza normalizzazione le performance peggioravano

-utilizzo della funzione np.round per fare da threshold della probabilità restituita dalla funzione logistica, assumendo come **valore di threshold 0.5**.


```python
class LogisticR:
  def __init__(self, tol=0.0001,max_iter=10000,alpha=0.001):
    self.tol=tol # Tolleranza per la convergenza dell'algoritmo di discesa del gradiente
    self.max_iter=max_iter # Numero massimo di iterazioni per l'algoritmo di discesa del gradiente
    self.alpha=alpha # Tasso di apprendimento (learning rate)
    self.weights=None # Pesi del modello
  def __Gradient_Descent(self,x,y):
    self.weights=np.zeros(x.shape[1]) # Inizializzazione dei pesi a zero
    for i in range(self.max_iter):
      h_x=1/(1+np.exp(-np.dot(x,self.weights))) # Calcolo dell'output del modello tramite la funzione sigmoide
      residui=y-h_x # Calcolo dei residui
      gradiente=np.dot(x.T,residui) # Calcolo del gradiente
      update=self.alpha/len(y)*gradiente # Calcolo dell'aggiornamento dei pesi
      self.weights+=update # Aggiornamento dei pesi del modello
      if (update<self.tol).all(): # Controllo di convergenza
        print("HO CONVERSO")
        return
  def fit(self, x, y):
    scaler = StandardScaler()
    x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
    x_array=x.to_numpy()
    x_array=np.insert(x_array,0,1,axis=1)
    if isinstance(y,pd.DataFrame):
      y=y["shares"]
    y_array=y.to_numpy()
    self.__Gradient_Descent(x_array,y_array) # Apprendimento dei pesi del modello utilizzando l'algoritmo di discesa del gradiente
  def predict(self,x):
    scaler = StandardScaler()
    x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
    x_array=x.to_numpy()
    x_array=np.insert(x_array,0,1,axis=1)
    if self.weights is None:
      raise Exception("Il modello va prima allenato")
    return np.round(1/(1+np.exp(-np.dot(x_array,self.weights)))) # Calcolo delle previsioni del modello tramite la funzione sigmoide
```

Implementazione K-NN.
Ho reso possibile cambiare da input **k**, iper-parametro che caratterizza il modello sia nella sua versione di **classificazione** che di **regressione** e la **task da eseguire**. Nel caso di classificazione effettua un **majority voting dei K sample più vicini** (mediante la funzione np.argmax) , nel caso di regressione la **media dei K sample più vicini** (mediante la funzione np.mean).

```python
class K_NN:
  def __init__(self,k=5,task="regression"):
    self.k=k
    self.task=task
    if self.task not in {"regression", "classification"}:
      raise Exception("task inammissibile: regression o classification")
    self.x=None
    self.y=None
  def fit(self, x, y):
    scaler = StandardScaler()
    x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
    x_array=x.to_numpy()
    if isinstance(y,pd.DataFrame):
      y=y["shares"]
    y_array=y.to_numpy()
    self.x=x_array
    self.y=y_array
  def predict(self, x):
    if self.x is None or self.y is None:
        raise Exception("Devi prima chiamare fit")
    scaler = StandardScaler()
    x = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
    y_pred = []
    x_array = x.to_numpy()
    insieme_distanze = np.zeros((x_array.shape[0], len(self.x)))
    for i in range(len(x_array)):
        distanze = np.sqrt(np.sum((self.x - x_array[i])**2, axis=1))
        neighbors_idx = np.argsort(distanze)[:self.k]
        neighbors_labels = self.y[neighbors_idx]
        if self.task=="classification":
          label = np.argmax(np.bincount(neighbors_labels))
        else:
          label = np.mean(neighbors_labels)
        y_pred.append(label)
    return y_pred
```

Implementazione reti neurali.
Per implementare la rete neurale ho trovato comodo implementare due classi tra loro collegate, una classe Layer caratterizzata da:
- **pesi w**
- **learning rate alpha**
-da una **funzione di attivazione**, da scegliere tra "logistic","relu","leaky-relu","linear"
- la sua **derivata**
-da a , l'**input del layer**
- da z, il **prodotto tra w e l'input del layer**.

Nella classe Neural Network ho reso disponibile scegliere l'input size (nel nostro caso sempre 58), una lista di interi hidden_layer_sizes che rappresenta il numero di nodi per hidden layer nella rete, funzione di attivazione, learning rate, numero massimo di iterazioni, tolleranza ed infine **la task naturalmente da scegliere tra "regression" e "classification"**.

Una nota particolarmente importante della implementazione è che oltre ad aver effettuato una **normalizzazione** ho anche effettuato un **clipping del gradiente** nel caso di classificazione, questo perchè altrimenti si ha durante l'esecuzione un overflow numerico.

Naturalmente, la differenza principale nella implementazione tra la task di classificazione e quella di regressione è l'utilizzo di una **Loss diversa (Binary Cross Entropy)** per determinarne la convergenza e della applicazione di una **threshold (impostata a 0.5) per la previsione delle classi**.

```python
class Layer:
  # Funzioni di attivazione disponibili e le loro derivate
  logistic = lambda x: 1/(1+np.exp(-x))
  d_logistic= lambda x: (1/(1+np.exp(-x)))*(1-1/(1+np.exp(-x)))
  relu=lambda x: np.maximum(0,x)
  d_relu=lambda x: x>0
  leaky_relu = lambda x : np.where(x > 0, x, x * 0.01)
  d_leaky_relu= lambda x :  np.where(x > 0, 1, 0.01)
  linear= lambda x: x #utilizzabile solo per l'output
  d_linear= lambda x: 1
  def __init__(self,input_nodes,output_nodes,alpha=0.0001,activation="relu",task="regression"):
    #inizializzazione di valori random per i pesi di un layer (He-et-Al)
    self.task=task
    self.w=np.random.randn(input_nodes+1,output_nodes)*np.sqrt(2/output_nodes)
    self.alpha=alpha
    if activation not in {"logistic","relu","leaky-relu","linear"}: #linear utilizzabile solo per l'output
      raise Exception("Attivazione deve essere tra le seguenti 3: logistic, relu, leaky-relu")
    # Assegnazione delle funzioni di attivazione e delle loro derivate
    if activation=="logistic":
      self.activation=Layer.logistic
      self.derivative=Layer.d_logistic
    elif activation=="relu":
      self.activation=Layer.relu
      self.derivative=Layer.d_relu
    elif activation=="leaky-relu":
      self.activation=Layer.leaky_relu
      self.derivative=Layer.d_leaky_relu
    else:
      self.activation=Layer.linear
      self.derivative=Layer.d_linear
    self.a=None
    self.z=None
  def forward(self,a):
    # Aggiunta del termine di bias
    self.a=np.insert(a,0,1,axis=1)
    if self.task=="classification":
      self.z=np.clip(self.a.dot(self.w),-100,100) #clipping per evitare overflow numerici
    else:
      self.z=self.a.dot(self.w)
    return self.activation(self.z)
  def back_propagation(self,output_error):
    delta =output_error * self.derivative(self.z)
    prev_error = delta.dot(self.w[1:].T)
    grad =self.a.T.dot(delta)
    self.w = self.w -self.alpha * grad
    return prev_error

class NeuralNetwork:
  def __init__(self, input_size, hidden_layer_sizes=[100], activation="relu",alpha=0.0001,max_iter=1000,tol=0.00001, task="regression"):
    if task not in{"classification","regression"}:
      raise Exception("INSERIRE TASK TRA CLASSIFICATION E REGRESSION")
    self.task=task
    self.layers=[]
    self.alpha=alpha
    # Creazione dei layer nascosti assegnando le funzioni di attivazione (linear per l'output della regressione e logistic per l'output della classificazione)
    for i in range(len(hidden_layer_sizes)):
      if i==0:
        self.layers.append(Layer(input_nodes=input_size,output_nodes=hidden_layer_sizes[i],alpha=alpha,activation=activation,task=task))
      else:
        self.layers.append(Layer(hidden_layer_sizes[i-1],hidden_layer_sizes[i],alpha=alpha,activation=activation,task=task))
    if task=="regression": #se la task è di regressione
      self.layers.append(Layer(hidden_layer_sizes[-1],1,alpha=alpha,activation="linear",task=task)) #ultimo layer attivazione linear
    else: #se task di classificazione
      self.layers.append(Layer(hidden_layer_sizes[-1],1,alpha=alpha,activation="logistic",task=task)) #ultimo layer attivazione logistic
    self.max_iter=max_iter #numero massimo di iterazioni
    self.tol=tol #tolleranza per la convergenza
  def __predict(self, x_t):
    #forwarding attraverso tutti i layers
    output=x_t
    for layer in self.layers:
      output=layer.forward(output)
    return output
  def fit(self,x_train,y_train):
    scaler = StandardScaler()
    x_train = pd.DataFrame(scaler.fit_transform(x_train), columns=x_train.columns)
    x_array=x_train.to_numpy()
    if isinstance(y_train,pd.DataFrame):
      y_train=y_train["shares"]
    y_array=y_train.to_numpy()
    prev_error=0
    for i in range(self.max_iter): #per ogni epoch
      error=0
      for (x,y) in zip(x_array,y_array): #per ogni sample del training set
        x=x.reshape(1,x.size)
        h_x=self.__predict(x) #prevedi l'output
        if self.task=="regression": #se la task è di regressione
          error+=1/len(y_array)*(y-h_x)**2 #calcolo MSE (verrà calcolata la radice) per la loss
        else: #se la task è di classificazione
            error +=((-y) * math.log(h_x+1e-6) - (-y+1) * math.log(1-h_x+1e-6))/(len(y_array)) #calcolo della binary cross entropy come loss
        delta=(h_x-y)/len(y_array) if self.task=="regression" else (h_x-y) #calcolo della differenza tra previsioni e ground truth per valutare l'errore, nel caso di classificazione è necessario dividere per len(y_array)
        for layer in self.layers[::-1]: #effettua la backpropagation tra i layers
          delta=layer.back_propagation(delta)/len(y_array) if self.task=="regression" else layer.back_propagation(delta) #se regressione necessario dividere il risultato della backpropagation su un layer per la len(y_array)
      if abs(error-prev_error)<self.tol: #criterio di convergenza
        print("HO CONVERSO")
        break
      print("Loss alla iterazione",i,math.sqrt(error)) #loss del modello
      prev_error=error #memorizza l'errore della iterazione corrente per usarla nella successiva iterazione
  def predict(self,x_test):
    scaler = StandardScaler()
    x_test = pd.DataFrame(scaler.fit_transform(x_test), columns=x_test.columns)
    x_array=x_test.to_numpy()
    y_pred=np.zeros(len(x_array))
    for i in range(len(x_array)):
      x=x_array[i].reshape(1,x_array[i].size)
      if self.task=="regression": #se la task è di regressione, fai forwarding dell'input per ottenere dalla rete una previsione
        y_pred[i]=self.__predict(x)
      else: #se la task è di regressione, fai forwarding dell'input e usa una treshold (0.5) per determinare la classe del sample
        prediction=round(self.__predict(x)[0][0],3)
        y_pred[i] = 1 if float(prediction)> 0.5 else 0
    return y_pred
```
